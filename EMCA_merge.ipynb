```python
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import Polygon
from dbfread import DBF
import jenkspy
import json
from jenkspy import jenks_breaks
```

```python
#ruta cv
CV_ruta = r"C:\Users\UTIC\Documents\Alejandro\Docs\GIS\EMCA\CICLOVIAS_2024.gpkg"
```

```python
#Leer archivo base

CV_gdf = gpd.read_file(CV_ruta, layer='CV_24_ajuste')
CV_POL_gdf = gpd.read_file(CV_ruta, layer='CAMBIO_CV_22_23')
```

```python
# Agrupar las localidades urbanas por municipio y sumar las áreas
long_cv = CV_gdf.dissolve(by='CLAVE_CV', aggfunc='sum')

# Renombrar la columna de área para claridad
long_cv.rename(columns={'DIST_KM': 'DIST_KM_TOT'}, inplace=True)
```

```python
# Unir la suma de áreas a la capa de delimitaciones municipales
CV_POL_gdf = CV_POL_gdf.merge(long_cv[['DIST_KM_TOT']], on='CLAVE_CV', how='left')
```

```python
CV_POL_gdf1 = CV_POL_gdf.drop(columns = ['DIST_KM', 'CALLE_1', 'CALLE_2', 
       'ANO_CONST', 'ANO_RCONST', 'GOB', 'AFO_2021', 'AFO_2020',
       'AFO_2022'])
```

```python
# Guardar el resultado en un nuevo archivo shapefile
CV_POL_gdf1.to_file(CV_ruta, layer='CAMBIO_CV_22_23_limpio', driver='GPKG')
```
