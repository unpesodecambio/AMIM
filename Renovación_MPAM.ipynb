```python
import pandas as pd

# Ruta de los archivos
ruta_original = r"C:\Users\AMIM GTAC\Desktop\mpam_completo.xlsx"
ruta_cajas = r"C:\Users\AMIM GTAC\Downloads\MPAM B.xlsx"
ruta_salida = r"C:\Users\AMIM GTAC\Downloads\MPAM B_nuevas.xlsx"

# Leer la lista maestra de mujeres
maestra = pd.read_excel(ruta_original, sheet_name='2024A')
nombres_maestra = set(maestra['Nombre'])  # Asegúrate de que el nombre de la columna sea correcto

# Leer el archivo de cajas
archivo_cajas = pd.ExcelFile(ruta_cajas)

# Hojas que queremos procesar
hojas = ['c8b', 'c9b', 'c10b', 'c11b', 'c12b', 'c13b', 'c14b', 'c15b', 'c16b']

# Crear un diccionario para almacenar los nuevos registros por hoja
nuevos_registros = {}

# Procesar cada hoja
for hoja in hojas:
    df = archivo_cajas.parse(hoja)  # Leer la hoja
    nombres_caja = set(df['Nombre'])  # Asume que la columna se llama "Nombre"
    nuevos_nombres = nombres_caja - nombres_maestra  # Identificar nuevos nombres
    nuevos_registros[hoja] = pd.DataFrame({'Nombre': list(nuevos_nombres)})  # Convertir a DataFrame

# Guardar el resultado en un nuevo Excel
with pd.ExcelWriter(ruta_salida) as writer:
    for hoja, df in nuevos_registros.items():
        df.to_excel(writer, sheet_name=hoja, index=False)

print(f"Archivo guardado en {ruta_salida}")
```
