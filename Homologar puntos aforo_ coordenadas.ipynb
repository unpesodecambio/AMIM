```python
import pandas as pd

# Cargar los datos desde el archivo Excel
file_path = r"C:\Users\UTIC\Documents\Alejandro\Docs\INFORME MOV\5. EMCA\Puntos AFO\PUNTOS AFORO 2019-2023.xlsx"

puntos_2020 = pd.read_excel(file_path, sheet_name='Puntos_2020')
puntos_2021 = pd.read_excel(file_path, sheet_name='Puntos_2021')
puntos_2022 = pd.read_excel(file_path, sheet_name='Puntos_2022')
puntos_2023 = pd.read_excel(file_path, sheet_name='Puntos_2023')
```

```python
def agregar_coordenadas(df_early, df_late):
    # Crear un DataFrame vacío para almacenar los resultados
    df_result = df_early.copy()

    # Buscar coincidencias y agregar coordenadas
    for i, row in df_result.iterrows():
        match = df_late[(df_late['Calle 1'] == row['Calle 1']) & (df_late['Calle 2'] == row['Calle 2'])]
        if not match.empty:
            df_result.at[i, 'x'] = match.iloc[0]['x']
            df_result.at[i, 'y'] = match.iloc[0]['y']
    
    return df_result

# Aplicar la función a los datos de 2020 y 2021
puntos_2020_con_coordenadas = agregar_coordenadas(puntos_2020, puntos_2022_2023)
puntos_2021_con_coordenadas = agregar_coordenadas(puntos_2021, puntos_2022_2023)
```

```python
# Guardar los resultados en un nuevo archivo Excel
with pd.ExcelWriter(r"C:\Users\UTIC\Documents\Alejandro\Docs\INFORME MOV\PUNTOS AFORO 2019-2023_ajuste.xlsx") as writer:
    puntos_2020_con_coordenadas.to_excel(writer, sheet_name='Puntos_2020', index=False)
    puntos_2021_con_coordenadas.to_excel(writer, sheet_name='Puntos_2021', index=False)
    puntos_2022.to_excel(writer, sheet_name='Puntos_2022', index=False)
    puntos_2023.to_excel(writer, sheet_name='Puntos_2023', index=False)
```

```python
from geopy.geocoders import Nominatim
import time

# Inicializar el geocodificador
geolocator = Nominatim(user_agent="myGeocoder")

def obtener_coordenadas(Calle_1, Calle_2):
    location = geolocator.geocode(f"{Calle_1} y {Calle_2}")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Función para obtener coordenadas con un retardo para evitar sobrecarga de la API
def obtener_coordenadas_con_reintentos(Calle_1, Calle_2, intentos=3):
    for _ in range(intentos):
        lat, lon = obtener_coordenadas(Calle_1, Calle_2)
        if lat is not None and lon is not None:
            return lat, lon
        time.sleep(1)  # Esperar 1 segundo entre intentos
    return None, None
```

```python
def completar_coordenadas(df):
    for i, row in df.iterrows():
        if pd.isna(row['x']) or pd.isna(row['y']):
            lat, lon = obtener_coordenadas_con_reintentos(row['Calle 1'], row['Calle 2'])
            df.at[i, 'x'] = lat
            df.at[i, 'y'] = lon
    return df

# Aplicar la función a los DataFrames de 2020 y 2021
puntos_2020_con_coordenadas = completar_coordenadas(puntos_2020_con_coordenadas)
puntos_2021_con_coordenadas = completar_coordenadas(puntos_2021_con_coordenadas)
```
