# Edades entrevistados

```python
# Datos de ejemplo
data = {
    'Edades': ['16-18', '19-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '60+'],
    '2020': [10, 15, 16, 14, 10, 8, 6, 5, 7, 9],
    '2021': [10, 13, 17, 14, 9, 11, 7, 6, 6, 7],
    '2022': [15, 16, 12.3, 10, 9, 8, 8, 7, 5, 9],
    '2023': [6, 12, 20, 15, 12, 8, 7, 7, 7, 6]
}

df = pd.DataFrame(data)
df.set_index('Edades', inplace=True)
df = df.transpose()

# Agrupar edades
df_grouped = pd.DataFrame({
    '16-25': df[['16-18', '19-25']].sum(axis=1),
    '26-40': df[['26-30', '31-35', '36-40']].sum(axis=1),
    '41-55': df[['41-45', '46-50', '51-55']].sum(axis=1),
    '56+': df[['56-60', '60+']].sum(axis=1)
})
```

```python
df_grouped
```

```python
# Gráfico de barras apiladas
df_grouped.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')
plt.title('Distribución de edades de entrevistados (2020-2023)')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.legend(title='Edades')
plt.show()
```

```python
df_grouped.plot(kind='line', figsize=(12, 8), marker='o')
plt.title('Distribución de edades de entrevistados (2020-2023)')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.legend(title='Edades')
plt.show()
```

# Estudios

```python
import pandas as pd

# Datos de ejemplo de nivel de estudios
data_estudios = {
    'Nivel de Estudios': ['Ninguno', 'Primaria', 'Secundaria', 'Preparatoria', 'Técnico', 'Licenciatura', 'Maestría', 'Doctorado', 'Prefiero no responder'],
    '2020': [0, 18, 27, 26, 4, 19, 2, 1, 3],
    '2021': [1, 14, 26, 36, 0, 21, 2, 0, 0],
    '2022': [2, 15.3, 24.7, 21.7, 1.3, 27.3, 1.7, 0, 6],
    '2023': [1, 5, 24, 37, 5, 24, 1, 1, 2]
}

df_estudios = pd.DataFrame(data_estudios)
df_estudios.set_index('Nivel de Estudios', inplace=True)

# Eliminar la categoría 'Prefiero no responder'
df_estudios = df_estudios.drop('Prefiero no responder')

# Agrupar categorías
df_estudios.loc['Educación básica'] = df_estudios.loc[['Primaria', 'Secundaria']].sum()
df_estudios.loc['Medio superior'] = df_estudios.loc[['Preparatoria', 'Técnico']].sum()
df_estudios.loc['Posgrado'] = df_estudios.loc[['Maestría', 'Doctorado']].sum()

# Eliminar categorías agrupadas
df_estudios = df_estudios.drop(['Primaria', 'Secundaria', 'Preparatoria', 'Técnico', 'Maestría', 'Doctorado'])

# Normalizar los valores para que sumen 100
df_normalizado = df_estudios.div(df_estudios.sum(axis=0), axis=1) * 100
```

```python
# Trasponer el DataFrame para que los años sean el índice
df_normalizado_t = df_normalizado.transpose()
```
