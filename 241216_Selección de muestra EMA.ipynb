```python
import pandas as pd
import numpy as np
from scipy.stats import norm

def limpiar_cp(cp):
    if pd.isna(cp):
        return 0
    cp = str(cp).strip()
    cp = ''.join(filter(str.isdigit, cp))
    return int(cp) if cp else 0

# Cargar las bases de datos con encoding diferente sin especificar dtype
usuarios = pd.read_csv(r"C:\Users\UTIC\Documents\Alejandro\Docs\INFORME MOV\6. MIBICI\usuarios_2023.csv", 
                       encoding='latin1')
encuestas = pd.read_csv(r"C:\Users\UTIC\Documents\Alejandro\Docs\INFORME MOV\6. MIBICI\EMA\Encuesta_MiBici_2023_Todas_Las_Respuestas_Individuales_240131\CSV\Encuesta MiBici 2023.csv")

# Limpiar la columna de códigos postales
usuarios['Shipping Zip/postal Code'] = usuarios['Shipping Zip/postal Code'].apply(limpiar_cp)
encuestas['¿Cuál es el código postal de la colonia donde resides?'] = encuestas['¿Cuál es el código postal de la colonia donde resides?'].apply(limpiar_cp)

# Renombrar columnas para facilidad
usuarios.rename(columns={'Shipping Zip/postal Code': 'Codigo_Postal', 'Gender': 'Genero'}, inplace=True)
encuestas.rename(columns={'¿Cuál es el código postal de la colonia donde resides?': 'Codigo_Postal', 'Género': 'Genero'}, inplace=True)
```

```python
# Parámetros de la muestra
confianza = 0.95
margen_error = 0.05

# Calculo del tamaño de muestra con población finita
N = len(usuarios)
Z = norm.ppf(1 - (1 - confianza) / 2)
p = 0.5  # Proporción esperada (máxima varianza)
n = (Z**2 * p * (1 - p)) / (margen_error**2)
n = n / (1 + (n - 1) / N)
n = int(np.ceil(n))  # Redondear hacia arriba para asegurar suficiente muestra
```

```python
# Distribución por código postal y género en la base de usuarios
dist_usuarios = usuarios.groupby(['Codigo_Postal', 'Genero']).size().reset_index(name='Cuenta')
total_usuarios = dist_usuarios['Cuenta'].sum()

# Calcular el porcentaje de cada grupo
dist_usuarios['Proporcion'] = dist_usuarios['Cuenta'] / total_usuarios

# Calcular el número de encuestas necesarias por grupo sin redondear
dist_usuarios['Encuestas_Necesarias_Float'] = dist_usuarios['Proporcion'] * n

# Redondear las encuestas necesarias
dist_usuarios['Encuestas_Necesarias'] = dist_usuarios['Encuestas_Necesarias_Float'].round()

# Ajustar para que la suma sea igual al tamaño de muestra requerido
ajuste = n - dist_usuarios['Encuestas_Necesarias'].sum()
dist_usuarios = dist_usuarios.sort_values(by='Encuestas_Necesarias_Float', ascending=False)
```

```python
# Convertir ajuste a entero
ajuste = int(round(ajuste))

# Ajustar para que la suma sea igual al tamaño de muestra requerido
dist_usuarios = dist_usuarios.sort_values(by='Encuestas_Necesarias_Float', ascending=False)

for i in range(abs(ajuste)):
    index = i % len(dist_usuarios)
    if ajuste > 0:
        dist_usuarios.iloc[index, dist_usuarios.columns.get_loc('Encuestas_Necesarias')] += 1
    else:
        dist_usuarios.iloc[index, dist_usuarios.columns.get_loc('Encuestas_Necesarias')] -= 1

# Mostrar la distribución ajustada
print(dist_usuarios)
print("Total de encuestas necesarias:", dist_usuarios['Encuestas_Necesarias'].sum())
```

```python
# Definir el mapeo de género
genero_map = {
    'Masculino': 'MALE',
    'Femenino': 'FEMALE'
}

# Aplicar el mapeo y asignar "OTHER" para categorías no coincidentes
encuestas['Genero'] = encuestas['Genero'].map(genero_map).fillna('OTHER')
```

```python
# Inicializamos un DataFrame vacío para almacenar las encuestas seleccionadas
encuestas_filtradas = pd.DataFrame()

# Iteramos sobre cada grupo y seleccionamos las encuestas necesarias
for idx, row in dist_usuarios.iterrows():
    cp = row['Codigo_Postal']
    genero = row['Genero']
    encuestas_necesarias = int(row['Encuestas_Necesarias'])  # Convertimos a entero

    # Filtramos las encuestas del código postal y género específicos
    encuestas_filtradas_cp_genero = encuestas[(encuestas['Codigo_Postal'] == cp) & (encuestas['Genero'] == genero)]
    
    # Imprimir información para depuración
    print(f"Procesando CP: {cp}, Género: {genero}")
    print(f"Encuestas necesarias: {encuestas_necesarias}")
    print(f"Total encuestas disponibles para este grupo: {len(encuestas_filtradas_cp_genero)}")

    # Seleccionamos las encuestas de manera aleatoria (puedes cambiar este método)
    if encuestas_necesarias <= len(encuestas_filtradas_cp_genero):
        encuestas_seleccionadas = encuestas_filtradas_cp_genero.sample(encuestas_necesarias)
    else:
        encuestas_seleccionadas = encuestas_filtradas_cp_genero

    # Imprimir cuántas encuestas fueron seleccionadas
    print(f"Encuestas seleccionadas: {len(encuestas_seleccionadas)}")

    # Añadimos las encuestas seleccionadas al DataFrame final
    encuestas_filtradas = pd.concat([encuestas_filtradas, encuestas_seleccionadas], ignore_index=True)

# Mostrar la cantidad total de encuestas seleccionadas al final
print(f"Total de encuestas seleccionadas: {len(encuestas_filtradas)}")
```
