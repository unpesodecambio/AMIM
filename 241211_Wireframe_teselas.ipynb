```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata

# Read the shapefile
shapefile_path = r"C:\Users\UTIC\Desktop\imeplan\Insumos\5. PDM\Teselas graduadas.shp"
gdf = gpd.read_file(shapefile_path)

# Extract centroids and altitude (assuming 'Prome' is the altitude attribute)
coordinates = []
altitudes = []

for geom, z in zip(gdf.geometry, gdf['Prome']):
    centroid = geom.centroid
    x, y = centroid.x, centroid.y
    coordinates.append((x, y))
    altitudes.append(z)

# Convert to numpy arrays for plotting
coordinates = np.array(coordinates)
altitudes = np.array(altitudes)

# Normalize the Z values to reduce height
altitudes = (altitudes - altitudes.min()) / (altitudes.max() - altitudes.min())

# Create a grid for interpolation
grid_x, grid_y = np.mgrid[coordinates[:, 0].min():coordinates[:, 0].max():200j, coordinates[:, 1].min():coordinates[:, 1].max():200j]

# Interpolate the Z values on the grid
grid_z = griddata(coordinates, altitudes, (grid_x, grid_y), method='cubic')

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the wireframe
ax.plot_wireframe(grid_x, grid_y, grid_z, color='black', linewidth=0.1)

# Set labels (optional)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title('3D Wireframe Map')

# Show the plot
plt.show()
```
